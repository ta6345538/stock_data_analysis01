## Importing

```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2020,4,17)
```

### Reading GOOGLE stock

```python
google = web.DataReader('GOOGL','yahoo',start,end)
```

```python
google.head()
```

### Stock Visualization

```python
google['Open'].plot(label = 'GOOGLE Open price', figsize=(15,7))
google['Close'].plot(label = 'GOOGLE Close price')
google['High'].plot(label = 'GOOGLE High price')
google['Low'].plot(label = 'GOOGLE Low price')
plt.legend()
plt.title('Google Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

### Analyzing stock for some Car Companies

```python
start1 = datetime.datetime(2012,1,1)
end1 = datetime.datetime(2017,1,1)
```

```python
tesla = web.DataReader('TSLA','yahoo',start1,end1)
ford = web.DataReader('F','yahoo',start1,end1)
gm = web.DataReader('GM','yahoo',start1,end1)
```

### Market Cap

```python
tesla['Total Traded'] = tesla['Open'] * tesla['Volume']
ford['Total Traded'] = ford['Open'] * ford['Volume']
gm['Total Traded'] = gm['Open'] * gm['Volume']
```

### Moving Average

```python
gm['Open'].plot(label='No Moving Average', figsize=(15,7))
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA50'].plot(label='MA50')
gm['MA200'] = gm['Open'].rolling(200).mean()
gm['MA200'].plot(label='MA200')
plt.legend()
plt.show()
```

### Correlation & Scatter Matrix

```python
from pandas.plotting import scatter_matrix
```

```python
car_comp = pd.concat([tesla['Open'],ford['Open'],gm['Open']],axis=1)
car_comp.columns = ['Tesla Open', 'Ford Open', 'Gm Open']
```

```python
scatter_matrix(car_comp,figsize=(8,8),hist_kwds={'bins':50})
plt.show()
```

### Daily Percentage Change

```python
tesla['Return'] = (tesla['Close']/tesla['Close'].shift(1) - 1)
```

```python
tesla.head()
```

```python
ford['Return'] = (tesla['Close']/tesla['Close'].shift(1) - 1)
```

```python
ford.head()
```

```python
gm['Return'] = (tesla['Close']/tesla['Close'].shift(1) - 1)
```

```python
gm.head()
```

```python
ford['Return'].hist(bins=50)
plt.show()
```

```python
tesla['Return'].hist(bins=100,label='Tesla',alpha=0.5, figsize = (13,6))
gm['Return'].hist(bins=100,label='GM',alpha=0.5)
ford['Return'].hist(bins=100,label='Ford',alpha=0.5)
plt.legend()
plt.show()
```

```python
tesla['Return'].plot(kind='kde',label='Tesla', figsize=(15,6))
ford['Return'].plot(kind='kde',label='Ford')
gm['Return'].plot(kind='kde',label='GM')
plt.legend()
plt.show()
```

```python
box_df = pd.concat([tesla['Return'],gm['Return'],ford['Return']], axis= 1)
box_df.columns=['Tesla Return','GM Return','Ford Return']
box_df.plot(kind='box',figsize=(16,6))
```

```python
scatter_matrix(box_df,figsize=(8,8),hist_kwds={'bins':50}, alpha = 0.25)
plt.show()
```

### Cumulative Return

```python
tesla['Cumulative Return'] = (1+ tesla['Return']).cumprod()
ford['Cumulative Return'] = (1+ ford['Return']).cumprod()
gm['Cumulative Return'] = (1+ gm['Return']).cumprod()
```

```python
tesla.head()
```

```python
ford.head()
```

```python
gm.head()
```

```python
tesla['Cumulative Return'].plot(label='Tesla', figsize=(15,7))
ford['Cumulative Return'].plot(label='Ford')
gm['Cumulative Return'].plot(label='GM')
plt.title('Cumulative Return VS time')
plt.legend()
plt.show()
```
